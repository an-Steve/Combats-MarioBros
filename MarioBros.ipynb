## **â†’Affichage de toute les datasheet**

```python
import pandas as pd
from tabulate import tabulate

# Charger le dataset
df = pd.read_csv("mario_dataset.csv")

# Fonction pour afficher un titre
def afficher_section(titre):
    print("\n" + "="*len(titre))
    print(titre)
    print("="*len(titre))

# 1ï¸âƒ£ AperÃ§u des 5 premiÃ¨res lignes
afficher_section("AperÃ§u des 5 premiÃ¨res lignes")
print(tabulate(df.head(), headers='keys', tablefmt='fancy_grid', showindex=False))

# 2ï¸âƒ£ AperÃ§u des 5 derniÃ¨res lignes
afficher_section("AperÃ§u des 5 derniÃ¨res lignes")
print(tabulate(df.tail(), headers='keys', tablefmt='fancy_grid', showindex=False))

# 3ï¸âƒ£ Informations gÃ©nÃ©rales
afficher_section("Infos du DataFrame")
print(df.info())

# 4ï¸âƒ£ Colonnes du dataset
afficher_section("Colonnes du dataset")
for col in df.columns:
    print(f"- {col}")

# 5ï¸âƒ£ Statistiques descriptives
afficher_section("Statistiques descriptives")
print(tabulate(df.describe(), headers='keys', tablefmt='fancy_grid'))

# 6ï¸âƒ£ VÃ©rifier les valeurs manquantes
afficher_section("Valeurs manquantes par colonne")
print(tabulate(df.isnull().sum().reset_index().rename(columns={'index':'Colonne', 0:'Valeurs manquantes'}),
               headers='keys', tablefmt='fancy_grid', showindex=False))

# 7ï¸âƒ£ Taille du DataFrame
afficher_section("Taille du DataFrame")
print(f"Nombre de lignes : {df.shape[0]}")
print(f"Nombre de colonnes : {df.shape[1]}")

# 8ï¸âƒ£ Afficher toutes les lignes si dataset petit
# pd.set_option('display.max_rows', None)
# print(tabulate(df, headers='keys', tablefmt='fancy_grid', showindex=False))
```

**â†’ Le Combat de l'univers de Mario**

```python
import pandas as pd
from tabulate import tabulate
import random

# Charger le dataset
df = pd.read_csv("mario_dataset.csv")

# VÃ©rifier les colonnes
print("Colonnes disponibles :")
print(df.columns.tolist())

col_personnages = 'player_id'  # Colonne des personnages

# SÃ©lectionner 10 personnages alÃ©atoires
df_10 = df.sample(n=10, random_state=42).reset_index(drop=True)

# Afficher les 10 personnages jouables
def afficher_personnages():
    tableau = [[idx + 1, pers] for idx, pers in enumerate(df_10[col_personnages])]
    print(tabulate(tableau, headers=["NÂ°", "Personnage"], tablefmt="fancy_grid"))

# Choix d'un personnage
def choisir_personnage():
    afficher_personnages()
    while True:
        try:
            choix = int(input("Choisis le numÃ©ro d'un personnage (1-10) : "))
            if 1 <= choix <= 10:
                personnage = df_10.iloc[choix-1]
                return personnage
            else:
                print("âš ï¸ Veuillez choisir un numÃ©ro entre 1 et 10.")
        except ValueError:
            print("âš ï¸ EntrÃ©e invalide. Veuillez entrer un numÃ©ro.")

# Combat alÃ©atoire entre deux personnages
def combat(p1, p2):
    gagnant = random.choice([p1[col_personnages], p2[col_personnages]])
    print(f"\nCombat entre {p1[col_personnages]} et {p2[col_personnages]} :")
    print(f"ğŸ† Le gagnant est {gagnant} !")

# Jeu principal
print("Bienvenue dans le jeu de combat Mario !\n")
personnage1 = choisir_personnage()
personnage2 = choisir_personnage()
combat(personnage1, personnage2)

```
